```python
def aggregate(df: pd.DataFrame, aggregate_columns: list, aggregate_dict: dict) -> pd.DataFrame:
    
    df2 = df.copy()
    final_df = df2.groupby(aggregate_columns).agg(aggregate_dict)
    
    return final_df
```
